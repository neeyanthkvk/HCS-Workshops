```python
id_tag = "question-mini-list"
```

```python
import requests
page = requests.get("https://math.stackexchange.com/")
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup.prettify()
```

```python
questions = soup.find(id = id_tag)
question_summary = questions.find_all(class_='summary')
fq = question_summary[0]
print(fq.prettify())
```
